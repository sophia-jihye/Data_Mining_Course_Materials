```python
import os 
import pandas as pd
import numpy as np

from nltk import sent_tokenize
from nltk.tokenize.punkt import PunktSentenceTokenizer

from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
```

```python
data_filepath = 'C:\DATA\Data_Mining_Course_Materials\마스크 속성별 리뷰.csv'

save_filepath_format = 'C:\DATA\Data_Mining_Course_Materials\마스크_가격,디자인,사이즈_{}.csv'
```

```python
df = pd.read_csv(data_filepath)
df.drop(columns=[col for col in df.columns if 'Unnamed:' in col], inplace=True)
df
```

```python
def extract_sentences(reviews):
    sentences = []
    for review in reviews:
        try:
            sentences.extend(sent_tokenize(review))
        except:
            print(review)
    return sentences
```

```python
dfs = []
for aspect_name in ['가격', '디자인', '사이즈']:
    reviews = [review for review in df[aspect_name].values if review is not np.nan]
    sentences = extract_sentences(reviews)
    print('Total number of sentences for the {} aspect: {}'.format(aspect_name, len(sentences)))
    print(sentences[:3], '\n')
    
    sent_df = pd.DataFrame(sentences, columns=['text'])
    sent_df['label'] = aspect_name
     
    # Balanced dataset 구축
    dfs.append(sent_df.sample(1000))
    
sent_df = pd.concat(dfs)
sent_df
```

```python
save_filepath = save_filepath_format.format(len(sent_df))
sent_df.to_csv(save_filepath, index=False)
print('Created {}'.format(save_filepath))
```
